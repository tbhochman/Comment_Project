```python
import csv

description = "In the first semester, we covered the fundamentals of Python, including loops, conditionals, functions, and dictionaries. In addition, we explored connections between math and computer science, including plotting, simulating the game of craps, and optimizing customer service at a bank."
students = [] #create empy lists for all the data in teacher.csv
last_names = []
genders = []
reflections = []
problem_sets = []
quizzes = []
assignments = []
final_exams = []
bank_projects = []
final_grades = []
fun = []

with open('teacher.csv') as csvfile: #open the csv and append each row to a separate list
    teacher = csv.reader(csvfile, delimiter=',')
    header = csvfile.readline()
    for row in teacher:
        #print(row)
        students.append(row[0])
        last_names.append(row[1])
        genders.append(row[2])
        reflections.append(row[-1])
        final_grades.append(row[3])
        problem_sets.append(row[6])
        quizzes.append(row[5])
        assignments.append(row[8])
        final_exams.append(row[4])
        bank_projects.append(row[7])
        fun.append(row[9])

        
def get_pronoun(gender): 
    '''returns the student's prefered pronoun given their gender'''
    if gender == 'M':
        return 'he'
    if gender == 'F':
        return 'she'
    else:
        return 'they'

def get_possessive(gender):
    '''returns the student's prefered pronoun in possessive form given their gender'''
    if gender == 'M':
        return 'his'
    if gender == 'F':
        return 'her'
    else:
        return 'their'

def get_pasttense(gender):
    '''returns verb conjugation of was/were given their gender'''
    if gender == 'M' or gender == 'F':
        return 'was'
    else:
        return 'were'

def get_object(gender):
    '''returns object form given their gender'''
    if gender == 'M':
        return 'him'
    if gender == 'F':
        return 'her'
    else:
        return 'them'
    
def get_overallperformance(i,student):
    '''returns the paragraph for overall performance given the student's final grade in the class'''
    if int(final_grades[i]) >= 90:
        return f"I am pleased with {student}’s progress this semester. As such, {possessive} problem set average was {problem_sets[i]}%, which indicates that {pronoun} had a solid understanding of the majority of the topics covered this semester and that {pronoun} {past_tense} consistent about meeting deadlines. For {possessive} quiz average and final exam score, {student} got {quizzes[i]}% and {final_exams[i]}%, respectively, which demonstrated that {pronoun} grasped the new content thoroughly and could hand write algorithms without using outside resources.\n"
    elif int(final_grades[i]) >= 80: 
        return f"I appreciate {student}’s energy this semester. While it is clear that Computer Science was perhaps not {possessive} preferred class choice, {pronoun} still performed well throughout the semester. As such, {student}'s problem set average was {problem_sets[i]}%, which indicates that {pronoun} had a decent understanding of the majority of the topics covered this semester and that {pronoun} {past_tense} somewhat consistent about meeting deadlines. For {possessive} quiz average and final exam score, {student} got {quizzes[i]}% and {final_exams[i]}%, respectively, which demonstrated that {pronoun} satisfactorily grasped the new content thoroughly and could often hand write algorithms without using outside resources.\n"
    else:
        return f"Admittedly, {student} struggled this semester, but {possessive} perseverance despite Computer Science not being {possessive} favorite class was admirable. As such, {possessive} problem set average was {problem_sets[i]}%, which indicates that {pronoun} had difficulty understanding some of the topics covered this semester and that {pronoun} {past_tense} sometimes inconsistent when it came to meeting deadlines. For {possessive} quiz average and final exam score, {pronoun} got {quizzes[i]}% and {final_exams[i]}%, respectively, which demonstrated that {pronoun} did not always grasp the new content thoroughly.\n"

def get_dailywork(i,student):
    '''returns the paragraph for daily work given the student's assignment grade average in the class'''
    if int(assignments[i]) >= 95:
        return f"On the daily coding activities, {student} worked mostly by {direct_object}self since the material came easily to {direct_object}, but {pronoun} was happy to help a classmate when they got stuck. In regard to using available resources like StackOverflow, {pronoun} wisely utilized these resources to troubleshoot and gain inspiration and internalized the solutions that {pronoun} found. In terms of coming in for extra help, {student} rarely, if ever, needed to do so because {pronoun} was so easily able to pick up the concepts in class.\n"
    elif int(assignments[i]) >= 90:
        return f"On the daily coding activities, {student} generally worked collaboratively since {pronoun} mostly understood the material, but it did not always come easily. Often, {pronoun} would help a classmate when they got stuck, or vice versa. In regard to using available resources like StackOverflow, {student} occasionally utilized these resources to troubleshoot and gain inspiration and internalized the solutions that {pronoun} found. In terms of coming in for extra help, {pronoun} sometimes came in for support, but I wish {pronoun} would have done so more often as it may have taken {direct_object} to the next level.\n"
    else:
        return f"On the daily coding activities, {student} mostly worked collaboratively since {pronoun} did not always pick up the material as easily as others. Often, {pronoun} would seek help from a classmate when {pronoun} got stuck. In terms of coming in for extra help, {student} occasionally came in for support, but I wish {pronoun} would have done so more often as it may have improved {possessive} success in the class.\n"
    
def get_bankingproject(i,student):
    '''returns the paragraph for the banking project given the student's bank project grade'''
    if int(bank_projects[i]) >= 95:
        return f"{student}’s group Bank Project grade was a {bank_projects[i]}%, which indicated that {possessive} effort, time management, and collaboration relating to larger group assignments were excellent. In addition, {pronoun} demonstrated significant enthusiasm and dedication when participating in the Hackathons. These assignments allowed students to put forth as much effort as they wished to, and {student} chose to use these opportunities to seriously challenge {direct_object}self, think creatively, and advance {possessive} skills.\n"
    elif int(bank_projects[i]) >= 90:
        return f"{student}’s group Bank Project grade was a {bank_projects[i]}%, which indicated that {possessive} effort, time management, and collaboration relating to larger group assignments were decent, although towards the end of the semester {pronoun} seemed to occasionally be struggling to concentrate. {student} demonstrated a relative degree of enthusiasm and dedication when participating in the Hackathons. These assignments allowed students to put forth as much effort as they wished to, and {student} generally chose to use these opportunities to challenge {direct_object}self and advance {possessive} skills.\n"
    else:
        return f"{student}’s group Bank Project grade was a {bank_projects[i]}%, which indicated that {possessive} effort, time management, and collaboration relating to larger group assignments were lacking. {student} also demonstrated a dearth of enthusiasm and dedication when participating in the Hackathons. These assignments allowed students to put forth as much effort as they wished to, and {student} generally chose to use these opportunities to rely on those around {direct_object} instead of trying to advance {possessive} skills.\n"

def get_fun(i,students):
    '''returns the paragraph for how fun the student is given their fun level'''
    if int(fun[i]) == 3:
        return f"Overall, {student}, without a doubt, had unbelievable amounts of fun in this class. More often than not, {pronoun} could be found wearing a pink unicorn uni-suit, humming to {direct_object}self as {pronoun} typed away. I’m not sure if it’s the cannabis legalization or the vibrations in the classroom, but {student} was just… ultra stoked on life this semester.\n"
    if int(fun[i]) == 2:
        return f"Overall, {student} seemed to generally enjoy {direct_object}self in this class. Generally, {pronoun} was upbeat as {pronoun} typed away. I hope that {student} enjoyed {possessive} first semester of Comp Sci!\n"
    else:
        return f"Ok. One last thing. Let me lay this out for you. If {student} were in Winnie the Pooh, {pronoun}’d be Eeyore. For hours on end, it seemed, {pronoun} would sit forlornly in the corner staring out the window imagining what I can only guess is the most efficient path towards their own defenestration. {student}'s verbal inflection was painfully monotonal, as it seemed that all vestiges of happiness and positive mental dispositions had been torn out of {possessive} existence. I really have nowhere to point {student}, as it seems that {pronoun} is destined to live out {possessive} days as a hermit in some far off abandoned island as far away from human contact as {pronoun} can possibly get.\n"
    
for i, student in enumerate(students): #loops through students 
    print(f"{student} {last_names[i]}:\n") #prints out the full name of the student
    with open(student.lower()+'.txt', 'w') as f: #creates a .txt for each student
        pronoun = get_pronoun(genders[i]) #stores the student's pronoun
        possessive = get_possessive(genders[i]) #stores the student's pronoun in possessive form
        past_tense = get_pasttense(genders[i]) #stores the past tense for the student's pronoun
        direct_object = get_object(genders[i]) #stores the object form of the student's pronoun
        f.write(description+'\n') #writes the description of the course into the comments
        f.write(get_overallperformance(i,student)) #writes the overall performance of the student
        f.write(f"In {student}'s self reflection, {pronoun} wrote: \n\t{reflections[i]}\n") #writes student's self reflection into comment
        f.write(get_dailywork(i,student)) #writes daily work paragraph for student
        f.write(get_bankingproject(i,student)) #writes paragraph about banking project for student
        f.write(get_fun(i,student)) #writes the paragraph for how fun the student was
        f.close()

    for line in open(student.lower()+'.txt'):
        print(line) #prints each student's file
```
